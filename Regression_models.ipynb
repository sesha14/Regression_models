```python
# Sesha Sai Sreevani Kappagantula


import pandas as pd
import seaborn as sns
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#load the dataset and print correlation matrix
boston_dataset=load_boston()
boston=pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
boston['MEDV']=boston_dataset.target
correlation_matrix=boston.corr().round(2)
sns.heatmap(data=correlation_matrix,annot=True)

#print correlation graph1
plt.figure(figsize=(20, 5)) 
features = ['LSTAT', 'RM']
target = boston['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
   
#print correlation graph2    
plt.figure(figsize=(20, 5))    
features = ['AGE', 'DIS']
target = boston['NOX']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('NOX')    

#print correlation graph3    
plt.figure(figsize=(20, 5))
features = ['INDUS', 'NOX']
target = boston['TAX']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = boston[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('TAX')  
    
#take data for regression    
train1 = boston.drop(['MEDV'],axis=1)
labels=boston['MEDV']

#use regressor
reg = LinearRegression()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train1, labels, test_size=0.2, random_state=1)
X_train1, X_val, y_train1, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
reg.fit(X_train1,y_train1)
k=reg.predict(X_train)
l=reg.predict(X_test)

print("\nvalidation score")
print(reg.score(X_val,y_val))
print("\ntest score")
print(reg.score(X_test,y_test))

rmse=(np.sqrt(mean_squared_error(y_train,k)))
print("\nrmse for train")
print(rmse)
r2 = r2_score(y_train, k)
print("\nr2_score for train")
print(r2)

rmse=(np.sqrt(mean_squared_error(y_test,l)))
print("\nrmse for test")
print(rmse)
r2 = r2_score(y_test, l)
print("\nr2_score for test")
print(r2)

print("\n\n")
#take data for regression    
boston['LSTAT']=boston_dataset.target
train1 = boston.drop(['LSTAT'],axis=1)
labels=boston['LSTAT']

#use regressor
reg = LinearRegression()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train1, labels, test_size=0.2, random_state=1)
X_train1, X_val, y_train1, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
reg.fit(X_train1,y_train1)
k=reg.predict(X_train)
l=reg.predict(X_test)

print("\nvalidation score")
print(reg.score(X_val,y_val))
print("\ntest score")
print(reg.score(X_test,y_test))

rmse=(np.sqrt(mean_squared_error(y_train,k)))
print("\nrmse for train")
print(rmse)
r2 = r2_score(y_train, k)
print("\nr2_score for train")
print(r2)

rmse=(np.sqrt(mean_squared_error(y_test,l)))
print("\nrmse for test")
print(rmse)
r2 = r2_score(y_test, l)
print("\nr2_score for test")
print(r2)
```
